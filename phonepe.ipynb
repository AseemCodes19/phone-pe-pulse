```python
#aggre_insurance
path1= "C:/Users/USER/Desktop/phonepe project/phonepe_pulse/data/aggregated/insurance/country/india/state/"
aggre_insurance_list=os.listdir(path1)

columns1={"state":[], "year":[], "quarter":[], "transaction_type":[], "transaction_count":[], "transaction_amount":[] }

for state in aggre_insurance_list:
    current_states=path1+state+"/"
    aggre_year_list=os.listdir(current_states)

    for year in aggre_year_list:
        current_year=current_states+year+"/"
        aggre_file_list=os.listdir(current_year)
      
        for file in aggre_file_list:
            current_file=current_year+file
            data=open(current_file,"r")

            A=json.load(data)

            for i in A ["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                columns1["transaction_type"].append(name)
                columns1["transaction_count"].append(count)
                columns1["transaction_amount"].append(amount)
                columns1["state"].append(state)
                columns1["year"].append(year)
                columns1["quarter"].append(int(file.strip(".json")))

    aggre_insurance=pd.DataFrame(columns1)

aggre_insurance["state"]= aggre_insurance["state"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
aggre_insurance["state"]= aggre_insurance["state"].str.replace("-","")
aggre_insurance["state"]= aggre_insurance["state"].str.title()
aggre_insurance["state"]= aggre_insurance["state"].str.replace('Dadra&Nagarhaveli&Daman&Diu','dadra and nagar haveli and daman and diu')
```
